```python
import numpy as pn
import pandas as pd
df=pd.read_csv('Life Expectancy Data.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
df.dropna(inplace=True)
df.duplicated().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
from sklearn.feature_selection import VarianceThreshold
numeric_df=df.select_dtypes(include='number')
constant_filter=VarianceThreshold(threshold=0)
constant_filter.fit(numeric_df)
constant_cols=[column for column,var in zip(numeric_df.columns, constant_filter.variances_) if var == 0]
constant_cols
```

```python
low_var=VarianceThreshold(threshold=0.01)
low_var.fit(numeric_df)
low_var=[column for column,var in zip(numeric_df.columns, low_var.variances_) if var<0.01]
low_var
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Country','Status'], drop_first=True)
```

```python
corr=df.corr()
target_corr=corr['Life expectancy'].sort_values(ascending=False)
print(target_corr.head(10))
```
