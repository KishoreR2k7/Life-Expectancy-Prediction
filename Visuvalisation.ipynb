```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Cleaned Data.csv')
df.head()
```

```python
df.columns
```

```python
sns.kdeplot(x=df['Life expectancy'],fill=True)
```

```python
features=df[["Schooling","Income composition of resources","BMI","Alcohol","Adult Mortality","HIV/AIDS","thinness  1-19 years","Life expectancy"]]
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(features.corr(),annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
features=["Schooling", "Income composition of resources", "BMI"]
```

```python
for col in features:
    plt.figure(figsize=(6,4))
    sns.kdeplot(df[col],fill=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```
